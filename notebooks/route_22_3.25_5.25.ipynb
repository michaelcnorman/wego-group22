```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import requests
import geojson
import json


%matplotlib inline
```

```python
with open('../data/WeGo.json') as fi:
    credentials = json.load(fi)
```

```python
Authorization = credentials['Authorization']
```

```python
URL = 'https://api.goswift.ly/speed-map/nashville/route/22'

params = {
    'direction': '1',
    'startDate': '03-25-2022',
    'beginTime' : '06:00',
    'resolution': 'hiRes',
}
params_geo  = {
    'direction': '1',
    'startDate': '03-25-2022',
    'endDate' : '04-24-2022',
    'format'   : 'geojson',
    'daysOfWeek': '1,5',
    'beginTime' : '06:00',
    'endTime' : '09:00',
    'resolution': 'hiRes'
}

headers = {'Authorization': Authorization,
          'Content-Type': 'application/json'}
```

```python
response_geo = requests.get(URL, params = params_geo, headers = headers)
```

```python
res = response_geo.json()['features']
res
```

```python
route22_gdf = gpd.GeoDataFrame.from_features(res)
route22_gdf.head()
```

```python
route22_gdf.sort_values(by='averageTravelSeconds',ascending=False)
```
