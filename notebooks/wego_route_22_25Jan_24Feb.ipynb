```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import pandas as pd
```

```python
import requests
import geojson
import json
```

```python
with open('../data/WeGo.json') as fi:
    credentials = json.load(fi)
```

```python
Authorization = credentials['Authorization']
```

```python
URL = 'https://api.goswift.ly/speed-map/nashville/route/22'

params_JantoFeb_mondayfriday = {
    'direction': '1', 
    'startDate': '01-25-2022',
    'endDate': '02-24-2022',
    'daysOfWeek': '1,5',
    'beginTime':'06:00',
    'endTime': '09:00',
    'resolution': 'hiRes',
} 
params_geo_JantoFeb_mondayfriday= {
    'direction': '1',
    'startDate': '01-25-2022',
    'endDate' : '02-24-2022',
    'format'   : 'geojson',
    'daysOfWeek': '1,5',
    'beginTime' : '06:00',
    'endTime' : '09:00',
    'resolution': 'hiRes'

}

params_JanFeb_TustoThru = {
    'direction': '1', 
    'startDate': '01-25-2022',
    'endDate': '02-24-2022',
    'daysOfWeek': '2,3,4',
    'beginTime':'06:00',
    'endTime': '09:00',
    'resolution': 'hiRes',
}

params_geo_JanFeb_TustoThru  = {
    'direction': '1',
    'startDate': '01-25-2022',
    'endDate' : '02-24-2022',
    'format'   : 'geojson',
    'daysOfWeek': '2,3,4',
    'beginTime' : '06:00',
    'endTime' : '09:00',
    'resolution': 'hiRes'
}

headers = {'Authorization': Authorization,
          'Content-Type': 'application/json'}
```

```python
response_geo1 = requests.get(URL, params = params_geo_JantoFeb_mondayfriday, headers = headers)
result_geo1 =response_geo1.json()['features']
georout22_gdf=gpd.GeoDataFrame.from_features(result_geo1)
georout22_gdf
```

```python
response_geo2 = requests.get(URL, params = params_JantoFeb_mondayfriday , headers = headers)
result2 =response_geo2.json()
rout22b_df=pd.DataFrame(result2['data']['segments'])
rout22b_df
```

```python
rout22_inbound_concat = pd.concat([georout22_gdf,rout22b_df[['countForSegment','percent20Secs']]], axis = 1)
rout22_inbound_concat.head(2)
```

```python
response_geo11 = requests.get(URL, params = params_geo_JanFeb_TustoThru, headers = headers)
result_geo11 =response_geo11.json()['features']
georout22_Tu_thu_gdf=gpd.GeoDataFrame.from_features(result_geo11)
georout22_Tu_thu_gdf.head(2)
```

```python
response_geo22 = requests.get(URL, params = params_JanFeb_TustoThru , headers = headers)
result22 =response_geo22.json()
rout22b_tus_thu_df=pd.DataFrame(result22['data']['segments'])
rout22b_tus_thu_df.head(2)
```

```python
rout22_inbound_concat2 = pd.concat([georout22_Tu_thu_gdf,rout22b_tus_thu_df[['countForSegment','percent20Secs']]], axis = 1)
rout22_inbound_concat2.head(2)
```
