```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import requests
import geojson
import json


%matplotlib inline
```

```python
with open('../data/WeGo.json') as fi:
    credentials = json.load(fi)
```

```python
Authorization = credentials['Authorization']
```

```python
Authorization
```

```python
URL = 'https://api.goswift.ly/speed-map/nashville/route/22'
params_AprilMay_MandF_AMPeak = {
    'direction': '1',
    'startDate': '03-25-2022',
    'endDate': '04-24-2022',
    'daysOfWeek': '1,5',
    'resolution': 'hiRes',
    'beginTime' : '06:00',
    'endTime' : '09:00'
}
params_geo_AprilMay_MandF_AMPeak  = {
    'direction': '1',
    'startDate': '03-25-2022',
    'endDate': '04-24-2022',
    'format'   : 'geojson',
    'daysOfWeek': '1,5',
    'beginTime' : '06:00',
    'endTime' : '09:00',
    'resolution': 'hiRes'
}
params_AprilMay_TthruTh_AMPeak = {
    'direction': '1',
    'startDate': '03-25-2022',
    'endDate': '04-24-2022',
    'daysOfWeek': '2,3,4',
    'resolution': 'hiRes',
    'beginTime' : '06:00',
    'endTime' : '09:00'
}
params_geo_AprilMay_TthruTh_AMPeak  = {
    'direction': '1',
    'startDate': '03-25-2022',
    'endDate': '04-24-2022',
    'format'   : 'geojson',
    'daysOfWeek': '2,3,4',
    'beginTime' : '06:00',
    'endTime' : '09:00',
    'resolution': 'hiRes'
}
headers = {'Authorization': Authorization,
         'Content-Type': 'application/json'}
```

```python
AprilMay_MandF_AMPeak = requests.get(URL, params = params_AprilMay_MandF_AMPeak, headers = headers)
```

```python
response_AprilMay_MandF_AMPeak = AprilMay_MandF_AMPeak.json()
response_AprilMay_MandF_AMPeak
```

```python
response_AprilMay_MandF_AMPeak = AprilMay_MandF_AMPeak.json()['data']['segments']
```

```python
route22_AprilMay_MandF_AMPeak_df = pd.DataFrame(response_AprilMay_MandF_AMPeak)
route22_AprilMay_MandF_AMPeak_df.head(5)
```

```python
route22_AprilMay_MandF_AMPeak_df = route22_AprilMay_MandF_AMPeak_df[['countForSegment','percent20Secs']]
route22_AprilMay_MandF_AMPeak_df.head(5)
```

```python
geo_AprilMay_MandF_AMPeak = requests.get(URL, params = params_geo_AprilMay_MandF_AMPeak, headers = headers)
```

```python
geo_response_AprilMay_MandF_AMPeak = geo_AprilMay_MandF_AMPeak.json()['features']
```

```python
geo_route22_AprilMay_MandF_AMPeak_df = gpd.GeoDataFrame.from_features(geo_response_AprilMay_MandF_AMPeak)
geo_route22_AprilMay_MandF_AMPeak_df.head(5)
```

```python
full_geo_route22_AprilMay_MandF_AMPeak_df = pd.concat([geo_route22_AprilMay_MandF_AMPeak_df, route22_AprilMay_MandF_AMPeak_df], axis=1)
full_geo_route22_AprilMay_MandF_AMPeak_df.head(5)
```

```python
full_geo_route22_AprilMay_MandF_AMPeak_df['time_of_day']='6am - 9am AM Peak'
full_geo_route22_AprilMay_MandF_AMPeak_df['start_date']= '03-25-2022'
full_geo_route22_AprilMay_MandF_AMPeak_df['end_date']= '04-24-2022'
full_geo_route22_AprilMay_MandF_AMPeak_df['direction']= 'Outbound'
full_geo_route22_AprilMay_MandF_AMPeak_df.head(2)
```
